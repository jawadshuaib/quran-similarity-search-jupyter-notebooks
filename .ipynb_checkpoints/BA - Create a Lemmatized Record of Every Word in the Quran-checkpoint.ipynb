## Use Farasa to Lemmatize Arabic of the Qur'an

Check out the Farasa website here for more details: https://farasa.qcri.org/lemmatization/

```python
from farasa.stemmer import FarasaStemmer
```

Creator of FarasaStemmer python library suggests using interactive mode when dealing with small texts (for example in a loop). More information can be found here: https://github.com/MagedSaeed/farasapy

```python
stemmer = FarasaStemmer (interactive=True)
```

## Create a list of unique words in the Quran

Quran - Simple (Plain) including tatweel but removing all other options (like pause marks)

Source: https://tanzil.net/download/

```python
unique_words = set()
filename = 'data/quran-simple-plain.txt'
with open(filename) as file:
    for line in file:
        line = line.rstrip()
        for word in line.split():
            if (word != '' and word != ' '):
                unique_words.add (word.strip())

print (unique_words)
```

```python
len (unique_words)
```

## Lemmatize Quranic Words

Lemmatize every word. Create a list containing dictionary of lemmatized words and their quranic equivalent

```python
ls = []
for unique_word in unique_words:
    lemmatized_word = stemmer.stem (unique_word)
    ls.append ({ 'lemmatized_word': lemmatized_word, 'quranic_word': unique_word })
```

## Translate Arabic to English

https://cloud.google.com/translate

Read the following tutorial for Google Cloud Translate API Setup: https://console.cloud.google.com/translation/introduction?walkthrough_id=translation_api_v3_index&project=quran-8b1f8&returnUrl=%2Ftranslation%2Fdashboard%3Fwalkthrough_id%3Dtranslation_api_v3_index%26project%3Dquran-8b1f8&cloudshell=true
        
For setting up the project on your local machine, you need to create a service account key that sits in your local machine: https://thenextweb.com/news/developers-implement-google-translate-apps-syndication

You can run gcloud using this /google-cloud-sdk/install.sh from root folder: https://cloud.google.com/sdk/docs/install-sdk

```python
from google.cloud import translate
```

```python
import os
os.environ["GOOGLE_APPLICATION_CREDENTIALS"]="connection/key.json"
```

```python
client = translate.TranslationServiceClient()
project_id="quran-8b1f8"
location = "global"
parent = f"projects/{project_id}/locations/{location}"
```

```python
def translate_text(text=""):
    response = client.translate_text(
        request={
            "parent": parent,
            "contents": [text],
            "mime_type": "text/plain",
            "source_language_code": "ar",
            "target_language_code": "en",
        }
    )

    translatedWord = ''
    for translation in response.translations:
        translatedWord = translation.translated_text
        
    return translatedWord
```

```python
translate_text('وَتُنذِرَ بِهِۦ')
```
