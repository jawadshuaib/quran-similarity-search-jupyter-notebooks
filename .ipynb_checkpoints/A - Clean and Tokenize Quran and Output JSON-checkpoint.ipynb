```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.tokenize import word_tokenize
from nltk.stem.porter import PorterStemmer
porter_stemmer = PorterStemmer()
```

```python
test_string = 'O chewed—up you who have believed, when you contract a debt for a specified term, write it down. And let a scribe write [it] between you in justice. Let no scribe refuse to write as Allāh has taught him. So let him write and let the one who has the obligation [i.e., the debtor] dictate. And let him fear Allāh, his Lord, and not leave anything out of it. But if the one who has the obligation is of limited understanding or weak or unable to dictate himself, then let his guardian dictate in justice. And bring to witness two witnesses from among your men. And if there are not two men [available], then a man and two women from those whom you accept as witnesses - so that if one of them [i.e., the women] errs, then the other can remind her. And let not the witnesses refuse when they are called upon. And do not be [too] weary to write it, whether it is small or large, for its [specified] term. That is more just in the sight of Allāh and stronger as evidence and more likely to prevent doubt between you, except when it is an immediate transaction which you conduct among yourselves. For [then] there is no blame upon you if you do not write it. And take witnesses when you conclude a contract. Let no scribe be harmed or any witness. For if you do so, indeed, it is [grave] disobedience in you. And fear Allāh. And Allāh teaches you. And Allāh is Knowing of all things.'
```

```python
def clean_text (text):
    
    # initializing bad_chars_list
    bad_chars = ['(', ')','[', ']','˹','˺', ',', '.', ';', ':', '!', "*", "”", "“"]
 
    # using filter() to
    # remove bad_chars
    text = ''.join((filter(lambda i: i not in bad_chars,
                                  text)))
    return str(text).lower()
```

```python
clean_string = (clean_text (test_string))
print (clean_string)
```

```python
def tokenize_string (text):
    
    # First Word tokenization
    nltk_tokens = nltk.word_tokenize(text)
    #Next find the roots of the word
    # print (nltk_tokens)
    return ' '.join([porter_stemmer.stem(w) for w in nltk_tokens])
```

```python
tokenized_string = tokenize_string (test_string)
print (tokenized_string)
```

```python
import re
```

```python
def remove_articles(text):
  return re.sub('\s+(and|you|have|has|as|or|a|an|in|is|are|for|it|to|the)(\s+)', '\2', text)
```

```python
print (tokenize_string(remove_articles (clean_text(test_string))))
```

```python
all_stopwords = stopwords.words('english')
all_stopwords.remove('not')
all_stopwords.remove('no')
all_stopwords.remove('what')
all_stopwords.remove('why')
all_stopwords.remove('he')
all_stopwords.remove('him')
all_stopwords.remove('his')
all_stopwords.remove('himself')
all_stopwords.remove('she')
all_stopwords.remove('she\'s')
all_stopwords.remove('her')
all_stopwords.remove('hers')
all_stopwords.remove('herself')
```

```python
# text = "Nick likes to play football, however he is not too fond of tennis."
def nltk_remove_stop_words (text):    
    text_tokens = word_tokenize(clean_text(text))
    return tokenize_string ((" ").join([word for word in text_tokens if not word in stopwords.words()]))
```

```python
nltk_remove_stop_words (test_string)
```

```python
formatted_quran = []
with open('en.itani.txt') as f:
    lines = f.readlines()
    for line in lines:        
        v = line.strip().split('|')
        surah_number = int(v[0])
        aya_number = int(v[1])
        translation = v[2]
        
        translation_clean = ''
        translation_tokenized = ''
        
        translation_clean = clean_text(translation)
        translation_tokenized = tokenize_string(translation_clean)
        translation_without_stop_words = tokenize_string (remove_stopwords(translation_clean))
        
        verse = {
            'translation_id':456, 
            'surah_number':surah_number, 
            'aya_number': aya_number, 
            'translation': translation,            
            'translation_clean': translation_clean,
            'translation_tokenized': translation_tokenized,
            'translation_without_stop_words': translation_without_stop_words
            }
        formatted_quran.append(verse)
        
formatted_quran
```

```python
import pandas as pd
```

```python
df = pd.DataFrame.from_records(formatted_quran)
```

```python
# export dataframe
df.to_json(r'en.itani.formatted.json')
```
